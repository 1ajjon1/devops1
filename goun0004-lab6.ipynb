```python
from github import Github
import yaml
#used personal account, 
data = yaml.safe_load(open("goun0004-credentials.yml"))


user = data["creds"]["username"]
token = data["creds"]["token"]


g = Github(token)

repo = g.get_repo("1ajjon1/devops1")

#gets branches
lst = list(repo.get_branches())
print(lst)

#gets Pull requests
pulls = repo.get_pulls(state='open', sort='created', base='master')
for pr in pulls:
    print(pr.number)

#gets list of commits 
commits = list(repo.get_commits())
for i in commits:
    print(i)
```
